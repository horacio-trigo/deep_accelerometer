```python
import pandas as pd
from os import getcwd, listdir, mkdir
from os.path import join, exists
import threading
```

```python
data_location = join(getcwd(), "data")
xpt_location = join(data_location, "xpt")
```

```python
files = [(folder, file_location) for folder in listdir(xpt_location) for file_location in listdir(join(xpt_location, folder))]
files
```

```python
csv_location = join(data_location, "csv")
if not exists(csv_location):
    mkdir(csv_location)
```

```python
def convert(folder, name):
    frame = pd.read_sas(join(xpt_location, folder, name))
    frame.to_csv(join(csv_location, folder, f"{name[:-4]}.csv"), index=False)
```

```python
threads = []

for folder, name in files:
    if not exists(join(csv_location, folder)):
        mkdir(join(csv_location, folder))

    x = threading.Thread(target=convert, args=(folder, name))
    threads.append(x)
    x.start()
    
for thread in threads:
    thread.join()
```
